```python
def CostFunction (theta,X,Y):
    m = X.shape[0];
    J = 0;
    for n in range(0,m):
        h = theta[0] + theta[1]*X[n] ;
        J = J + pow((h - Y[n]),2);
    J = J/(2*m);
    return J;
```

```python
iterations = 20000;
alpha = 0.01;
theta = np.array([0.,0.]);
for s in range(0,iterations):
    m = X.shape[0];
    a = 0;
    for n in range(0,m):
        h = theta[0] + theta[1]*X[n] ;
        a = a + (h - Y[n]);
    a = a/(m);
    q = theta.shape[0];
    for n in range(0,q):
        theta[n] = theta[n] - alpha*a;
theta
```
